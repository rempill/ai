```python
import os
import cv2
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_lfw_people
```

```python
# 1. Pregătire directoare
os.makedirs('data/Original', exist_ok=True)
os.makedirs('data/Sepia', exist_ok=True)

# 2. Descărcare date (LFW)
lfw = fetch_lfw_people(min_faces_per_person=1, resize=1.0, color=True)
NR_IMAGINI = 1000

date_etichete = []

for i in range(NR_IMAGINI):
    # Imaginea Originală
    img = (lfw.images[i] * 255).astype(np.uint8)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    img = cv2.resize(img, (128, 128))

    nume_original = f"img_{i}.jpg"
    cale_original = f"data/Original/{nume_original}"
    cv2.imwrite(cale_original, img)
    date_etichete.append({'nume_fisier': nume_original, 'eticheta': 0, 'cale': cale_original})

    # Imaginea Sepia
    matrice_sepia = np.array([[0.272, 0.534, 0.131],
                             [0.349, 0.686, 0.168],
                             [0.393, 0.769, 0.189]])
    img_sepia = cv2.transform(img, matrice_sepia)
    img_sepia = np.clip(img_sepia, 0, 255).astype(np.uint8)

    nume_sepia = f"img_{i}_sepia.jpg"
    cale_sepia = f"data/Sepia/{nume_sepia}"
    cv2.imwrite(cale_sepia, img_sepia)
    date_etichete.append({'nume_fisier': nume_sepia, 'eticheta': 1, 'cale': cale_sepia})

# 3. Creare fișier de etichete (CSV)
df = pd.DataFrame(date_etichete)
df.to_csv('data/etichete.csv', index=False)
```
