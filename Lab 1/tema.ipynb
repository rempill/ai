```python
# ex 1 : my solution

def last_word(string):
    if not string:
        return ""
    words=string.split()
    if words:
        return max(words)
    return ""

# ex 1 : ai solution
def last_word_ai(text):
    if not text: return ""
    words = text.split()
    return max(words) if words else ""

# complexitate: identica, O(n) timp si O(n) spatiu
# cazuri extreme: string initial vid, string fara cuvinte

print(last_word("Ana are mere rosii si galbene"))
print(last_word_ai("Ana are mere rosii si galbene"))
```

```python
import math

# ex 2 : my solution
def dist_euclid(p1,p2):
    return math.sqrt((p2[0]-p1[0])**2+(p2[1]-p1[1])**2)

# ex 2: ai solution
def dist_euclid_ai(p1, p2):
    return math.sqrt(sum((a - b) ** 2 for a, b in zip(p1, p2)))

# complexitate eu: O(1) timp si spatiu
# complexitate ai: O(d) timp si O(1) spatiu, d e dimensiunea pct ( a generalizat pentru n dimensiuni)
# cazuri extreme: dimensiuni diferite ( a mea va da eroare, ai se adapteaza)

print(dist_euclid((1,5),(4,1)))
print(dist_euclid_ai((1,5),(4,1)))
```

```python
# ex 3 : my solution
def produs_scalar(v1,v2):
    return sum(a*b for a,b in zip(v1,v2) if a != 0 and b != 0)

# ex 3 : ai solution
def produs_scalar_ai(v1, v2):
    # Reprezentare rară: doar indicii nenuli
    dict1 = {i: val for i, val in enumerate(v1) if val != 0}
    res = 0
    for i, val in enumerate(v2):
        if val != 0 and i in dict1:
            res += val * dict1[i]
    return res

# complexitate eu: O(n) timp si O(1) spatiu ( zip face iterator )
# complexitate ai: O(n) timp si O(k) spatiu, k e numarul de elemente nenule in v1
# cazuri extreme: vectori de dim diferite ( zip() se opreste la lungimea minima, deci nu va da eroare), vectori cu multe elemente nenule ( ai va folosi mai mult spatiu)

print(produs_scalar([1,0,2,0,3],[1,2,0,3,1]))
print(produs_scalar_ai([1,0,2,0,3],[1,2,0,3,1]))
```

```python
from collections import Counter

# ex 4: my solution
def unique_words(string):
    words=string.split()
    freq={}
    for elem in words:
        freq[elem]=freq.get(elem,0)+1
    return [elem for elem in freq if freq[elem]==1]

# ex 4: ai solution
def unique_words_ai(text):
    words = text.split()
    counts = Counter(words)
    return [word for word in counts if counts[word] == 1]

# complexitate: O(n) timp si spatiu ambele.
# cazuri extreme: toate cuv identice ( lista goala ), toate cuv diferite ( lista cu toate cuvintele ), string vid ( lista goala )

print(unique_words("ana are ana are mere rosii ana"))
print(unique_words_ai("ana are ana are mere rosii ana"))
```

```python
# ex 5: my solution
def duplicate_num(lista):
    return sum(lista)-len(lista)*(len(lista)-1)//2

# ex 5: ai solution
def duplicate_num_ai(nums):
     n = len(nums)
     expected_sum = (n - 1) * n // 2
     return sum(nums) - expected_sum

# complexitate: ambele O(n) timp si O(1) spatiu
# cazuri extreme: nu exista

print(duplicate_num([1,2,3,4,2]))
print(duplicate_num_ai([1,2,3,4,2]))
```
