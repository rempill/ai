```python
# ex 1 : my solution

"""Returneaza cuvantul maxim (alfabetic) dintr-un text.
    Input:
      string (str): text cu cuvinte separate prin spatiu
    Output:
      str: cuvantul care este ultimul dpdv alfabetic
    """
def last_word(string):
    if not string:
        return ""
    words=string.split()
    if words:
        return max(words)
    return ""

# ex 1 : ai solution
def last_word_ai(text):
    if not text: return ""
    words = text.split()
    return max(words) if words else ""

# complexitate: identica, O(n) timp si O(n) spatiu (split)
# cazuri extreme: string initial vid, string fara cuvinte

print("me:",last_word("Ana are mere rosii si galbene"))
print("ai:",last_word_ai("Ana are mere rosii si galbene"))
```

```python
import math

"""Distanta euclidiana intre doua puncte 2D.
    Input:
      p1, p2: tuple/list cu 2 coordonate (x, y).
    Output:
      float: sqrt((diff x)^2 + (diff y)^2).
    """
# ex 2 : my solution
def dist_euclid(p1,p2):
    return math.sqrt((p2[0]-p1[0])**2+(p2[1]-p1[1])**2)

# ex 2: ai solution
def dist_euclid_ai(p1, p2):
    return math.sqrt(sum((a - b) ** 2 for a, b in zip(p1, p2)))

# complexitate eu: O(1) timp si spatiu
# complexitate ai: O(d) timp si O(1) spatiu, d e dimensiunea pct ( a generalizat pentru n dimensiuni)
# cazuri extreme: dimensiuni diferite ( a mea va da eroare, ai se adapteaza)

print("me:",dist_euclid((1,5),(4,1)))
print("ai:",dist_euclid_ai((1,5),(4,1)))
```

```python
# ex 3 : my solution
def produs_scalar(v1,v2):
    """Produs scalar pentru vectori rari (sparse) reprezentati ca liste.
    Input:
      v1, v2: iterabile de numere.
    Output:
      numar: suma(v1[i] * v2[i]) pentru i comun.
    Implementare: zip() merge pana la lungimea minima; ignoram inmultiri cu 0
    """
    return sum(a*b for a,b in zip(v1,v2) if a != 0 and b != 0)

# ex 3 : ai solution
def produs_scalar_ai(v1, v2):
    # Reprezentare rară: doar indicii nenuli
    dict1 = {i: val for i, val in enumerate(v1) if val != 0}
    res = 0
    for i, val in enumerate(v2):
        if val != 0 and i in dict1:
            res += val * dict1[i]
    return res

# complexitate eu: O(n) timp si O(1) spatiu ( zip face iterator )
# complexitate ai: O(n) timp si O(k) spatiu, k e numarul de elemente nenule in v1
# cazuri extreme: vectori de dim diferite ( zip() se opreste la lungimea minima, deci nu va da eroare), vectori cu multe elemente nenule ( ai va folosi mai mult spatiu)

print("me:",produs_scalar([1,0,2,0,3],[1,2,0,3,1]))
print("ai:",produs_scalar_ai([1,0,2,0,3],[1,2,0,3,1]))
```

```python
from collections import Counter

# ex 4: my solution
def unique_words(string):
    """Cuvintele care apar exact o singura data intr-un text.
    Input:
      string (str): text cu cuvinte separate prin spatiu.
    Output:
      list[str]: cuvinte cu frecventa 1.
    """
    words=string.split()
    freq={}
    for elem in words:
        freq[elem]=freq.get(elem,0)+1
    return [elem for elem in freq if freq[elem]==1]

# ex 4: ai solution
def unique_words_ai(text):
    words = text.split()
    counts = Counter(words)
    return [word for word in counts if counts[word] == 1]

# complexitate: O(n) timp si spatiu ambele.
# cazuri extreme: toate cuv identice ( lista goala ), toate cuv diferite ( lista cu toate cuvintele ), string vid ( lista goala )

print("me:",unique_words("ana are ana are mere rosii ana"))
print("ai:",unique_words_ai("ana are ana are mere rosii ana"))
```

```python
# ex 5: my solution
def duplicate_num(lista):
    """Gaseste valoarea duplicata cand lista are n elemente din {1..n-1} si exact una se repeta.

    Input:
      lista (list[int]): lungime n, contine 1..n-1, o valoare apare de 2 ori.
    Output:
      int: valoarea care se repeta.

    Idee: suma(lista) - suma(1..n-1) = duplicatul.
    """
    return sum(lista)-len(lista)*(len(lista)-1)//2

# ex 5: ai solution
def duplicate_num_ai(nums):
     n = len(nums)
     expected_sum = (n - 1) * n // 2
     return sum(nums) - expected_sum

# complexitate: ambele O(n) timp si O(1) spatiu
# cazuri extreme: nu exista

print("me:",duplicate_num([1,2,3,4,2]))
print("ai:",duplicate_num_ai([1,2,3,4,2]))
```

```python
# ex 6: my solution
def common_element(array):
    """Elementul majoritar (apare de > n/2 ori).

    Input:
      array (list[int]): lista.
    Output:
      int|None: candidatul majoritar.

    Algoritm: Boyer-Moore Voting
    """
    element=None
    counts=0
    for elem in array:
        if counts==0:
            element=elem
            counts=1
        elif elem==element:
            counts+=1
        else:
            counts-=1
    return element

# ex 6: ai solution
def common_element_ai(nums):
    candidate = None
    count = 0
    for num in nums:
        if count == 0:
            candidate = num
        count += (1 if num == candidate else -1)
    return candidate

# complexitate: ambele O(n) timp si O(1) spatiu
# cazuri extreme: toate elem identice ( va returna acel el ), lista cu un singur el ( va returna acel el ), lista goala ( va returna None )

print("me:",common_element([2,8,7,2,2,5,2,3,1,2,2]))
print("ai:",common_element_ai([2,8,7,2,2,5,2,3,1,2,2]))
```

```python
from heapq import heapify, heappush, heappop, nlargest

# ex 7: my solution
def kth_largest(array,k):
    """Al k-lea cel mai mare element dintr-un sir.

    Input:
      array (list[int/float]): sirul de valori.
      k (int): rangul.
    Output:
      valoarea k-th largest.

    Idee: tinem in min-heap doar cele mai mari k elemente vazute; radacina (min) e elementul cautat.
    """
    minheap=[]
    heapify(minheap)
    for elem in array:
        heappush(minheap,elem)
        if len(minheap)>k:
            heappop(minheap)  # eliminam cel mai mic din cele k+1, ca sa ramana doar top-k
    return minheap[0]

# ex 7: ai solution
def kth_largest_ai(nums, k):
    return nlargest(k, nums)[-1]

# complexitate: ambele O(n log k) timp si O(k) spatiu
# cazuri extreme: k>n ( ambele arunca error )

print("me:",kth_largest([7,4,6,3,9,1],2))
print("ai:",kth_largest_ai([7,4,6,3,9,1],2))
```

```python
# ex 8: my solution
def n_binary_nums(n):
    """Genereaza reprezentarile binare pentru numerele 1..n.

    Input:
      n (int): limita superioara.
    Output:
      list[str].
    """
    if n==0:
        return 0
    res=[]
    for i in range(1,n+1):
        bits=[]
        while i>0:
            bits.append(str(i%2))
            i//=2 # obtinem urmatorul bit prin impartire la 2; restul e bitul curent
        bits.reverse()
        res.append("".join(bits))
    return res

# ex 8: ai solution
def n_binary_nums_ai(n):
    return [bin(i)[2:] for i in range(1, n + 1)]

# complexitate: O(n log n) timp si O(n log n) spatiu ambele ( fiecare numar are cel mult log n biti )
# cazuri extreme: n=0 ( a mea returneaza 0, ai returneaza lista goala )

print("me:",n_binary_nums(4))
print("ai:",n_binary_nums_ai(4))
```

```python
# ex 9: my solution
def submatrix_sum(matrix,coords):
    """Suma elementelor pentru fiecare sub-matrice data.

    Input:
      matrix (list[list[int]]): matrice n x m.
      coords (list[tuple[tuple[int,int], tuple[int,int]]]): lista de query-uri.
    Output:
      list[int]: suma pe fiecare submatrice.

    Idee: Brute-Force.
    """
    sums=[]
    for (r1, c1), (r2, c2) in coords:
        s=0
        for r in range(r1,r2+1):
            for c in range(c1,c2+1):
                s+=matrix[r][c]
        sums.append(s)
    return sums

# ex 9: ai solution
class MatrixSum:
    def __init__(self, matrix):
        if not matrix: return
        rows, cols = len(matrix), len(matrix[0])
        self.sums = [[0] * (cols + 1) for _ in range(rows + 1)]
        for r in range(rows):
            for c in range(cols):
                self.sums[r+1][c+1] = matrix[r][c] + self.sums[r][c+1] + \
                                      self.sums[r+1][c] - self.sums[r][c]

    def get_sum(self, r1, c1, r2, c2):
        # r1, c1 (sus-stânga), r2, c2 (jos-dreapta)
        return self.sums[r2+1][c2+1] - self.sums[r1][c2+1] - \
               self.sums[r2+1][c1] + self.sums[r1][c1]

def submatrix_sum_ai(matrix, coords):
    ms=MatrixSum(matrix)
    return [ms.get_sum(r1, c1, r2, c2) for (r1, c1), (r2, c2) in coords]

# complexitate eu: O(M*N*Q) timp si O(1) spatiu, unde M,N dimensiunile matricei, Q numarul de interogari
# complexitate ai: O(M*N+Q) timp si O(M*N) spatiu, datorita matricei de sume prefixate
# cazuri extreme: matrice vida, matrice cu un singur element, interogari care acopera toata matricea ( ambele returneaza suma totala ), interogari care acopera un singur element ( ambele returneaza acel element )


print("me:",submatrix_sum([[0, 2, 5, 4, 1],
[4, 8, 2, 3, 7],
[6, 3, 4, 6, 2],
[7, 3, 1, 8, 3],
[1, 5, 7, 9, 4]],[((1, 1), (3, 3)), ((2,2), (4,4))]))
print("ai:",submatrix_sum_ai([[0, 2, 5, 4, 1],
[4, 8, 2, 3, 7],
[6, 3, 4, 6, 2],
[7, 3, 1, 8, 3],
[1, 5, 7, 9, 4]],[((1, 1), (3, 3)), ((2,2), (4,4))]))
```

```python
# ex 10: my solution
def most_1s(matrix):
    """Indexul liniei cu cele mai multe valori 1 intr-o matrice binara sortata crescator pe linii.

    Input:
      matrix (list[list[int]]): matrice cu 0/1.
    Output:
      int: indexul primei linii cu nr maxim de 1.

    Idee: cautam primul 1 pe coloane, de la stanga la dreapta.
    """
    if not matrix or not matrix[0]:
        return -1
    for col in range(len(matrix[0])):
        for i in range(len(matrix)):
            if matrix[i][col]==1:
                return i # prima linie care are 1 in aceasta coloana, restul de jos pot avea max la fel, dar nu mai mult
    return -1

# ex 10: ai solution
def most_1s_ai(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    max_row_idx = 0
    j = cols - 1 # Începem de la ultima coloană
    for i in range(rows):
        while j >= 0 and matrix[i][j] == 1:
            j -= 1
            max_row_idx = i
    return max_row_idx

# complexitate eu: O(N*M) timp si O(1) spatiu, unde N e numarul de linii, M numarul de coloane
# complexitate ai: O(N+M) timp si O(1) spatiu, datorita parcurgerii eficiente a matricei pornind din dreapta sus
# cazuri extreme: matrice vida ( al meu -1, ai eroare), matrice fara 1 ( al meu -1, ai 0 ), toate elemente 1 ( al meu 0, ai ultima linie ), linii cu nr egal de 1 ( al meu prima linie, ai ultima linie )

print("me:",most_1s([[0,0,0,1,1],
[0,1,1,1,1],
[0,0,1,1,1]]))
print("ai:",most_1s_ai([[0,0,0,1,1],
[0,1,1,1,1],
[0,0,1,1,1]]))
```
